## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

## Import datasets

```python
from yahoo_fin import stock_info
df = stock_info.get_data(input('Enter ticker: '))
df = df.reset_index()
df.rename(columns={'index': 'datetime'}, inplace=True)
df['datetime'] = df['datetime'].apply(lambda x: datetime.timestamp(x))
X = np.delete(df.iloc[:, 0:-2].values, obj=[2, 3, 4, 5], axis=1)
y = df.iloc[:, 4].values
```

## Splitting dataset for training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

## Training model

```python
from sklearn.ensemble import RandomForestRegressor
from datetime import datetime
model = RandomForestRegressor(n_estimators=5)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print('datetime\t\ty_test\ty_pred')
print('-'*38)
for i, j, k in zip(X_test[:, 0], y_test, y_pred):
    print(f'{datetime.fromtimestamp(i)}\t{round(j, 2)}\t{round(k, 2)}')
```

## Make prediction from user input

```python
model.predict([[datetime.timestamp(datetime.strptime(input("Enter date (dd/mm/yy): "), '%d/%m/%y')), float(input('Enter Open price: '))]])
```
